```python
#bring in dataframe
df = pd.read_csv(r"C:\Users\Burgundy\Documents\Data Analysis Projects\Project7-MovieData\MovieAndTVData.csv")
df
```

```python
#checking number of null values and data types for each column
df.info()
```

```python
#view non-null values for Gross columns to determine needed formatting
df[df[['Gross']].notna().all(axis=1)]
```

```python
#Remove not needed columns
df = df.drop(columns = ["STARS", "ONE-LINE", "VOTES"] )
df
```

```python
#Remove unnecessary duplicate values
df = df.drop_duplicates()
df
```

```python
#Create check point
```

```python
#Reformat Gross Column
df["Gross"] = df["Gross"].str.strip("$BM")
df[df[['Gross']].notna().all(axis=1)]
```

```python
#Change Gross column datatype to float
df['Gross'] = df['Gross'].astype('float')
df[df[['Gross']].notna().all(axis=1)]
```

```python
#Remove unwanted characters from Genre column
df["GENRE"] = df["GENRE"].str.strip("\n")
df
```

```python
#Remove end year from TV shows
df["YEAR"] = df["YEAR"].str.split("–").str[0]
df
```

```python
#Remove unwanted characters from Year column
df["YEAR"] = df["YEAR"].str.strip("()")
df
```
